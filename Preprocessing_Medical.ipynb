```python
import os
import shutil
a=["train", "val"]
for i in a:
    print(i)
    # Define the paths to the dataset, text file and the output directory
    dataset_path = os.path.join("/content/drive/MyDrive/dataset_divided/", str(i))
    print(dataset_path)
    text_file_path = "/content/drive/MyDrive/chestLANT.txt"
    output_directory = os.path.join("/content/drive/MyDrive/dataset_divided/",str(i))

    # Read the text file and store the image names and labels in a dictionary
    labels_dict = {}
    with open(text_file_path, 'r') as f:
        lines = f.readlines()
        for line in lines:
            image_name, label = line.strip().split()
            labels_dict[image_name] = label

    # Create folders for each label in the output directory
    for label in set(labels_dict.values()):
        label_folder_path = os.path.join(output_directory, label)
        os.makedirs(label_folder_path, exist_ok=True)

    # Move the images to their respective label folders
    for image_name, label in labels_dict.items():
        src_path = os.path.join(dataset_path, image_name)
        dst_path = os.path.join(output_directory, label, image_name)
        if os.path.exists(src_path):
          shutil.move(src_path, dst_path)
```

```python
import numpy as np
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from keras.preprocessing.image import ImageDataGenerator

# Set the path to the directory containing the bone scan images
train_dir = '/content/drive/MyDrive/Pre-processed/train'
test_dir = '/content/drive/MyDrive/Pre-processed/val'

# Set the batch size, image size, and number of classes
batch_size = 32
img_size = (256, 1024)
num_classes = 2

# Create an instance of the ImageDataGenerator class for data augmentation
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

# Create generators for loading the training and testing data
train_generator = datagen.flow_from_directory(
    train_dir,
    #target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical')

test_generator = datagen.flow_from_directory(
    test_dir,
    #target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical')

# Define the CNN model architecture
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(256, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(num_classes, activation='softmax'))

# Compile the model and specify the loss function, optimizer, and evaluation metric
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Train the model on the training data and evaluate on the testing data
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.n // batch_size,
    epochs=10,
    validation_data=test_generator,
    validation_steps=test_generator.n // batch_size)
# Save the model weights and architecture to disk
model.save_weights('cancer_diagnosis_weights.h5')
model.save('cancer_diagnosis_model.h5')
```
